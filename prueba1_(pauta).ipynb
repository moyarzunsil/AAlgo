3. **De la mejor a la peor, enumere las funciones recurrentes usadas en el análisis de algoritmos y dibújelas en un solo gráfico.**

**R:** Las funciones recurrentes de mejor a peor son:

- Constante ($O(1)$)
- Logarítmico ($O(\lg n)$)
- Lineal ($O(n)$)
- Lineal-Logarítmico ($O(n \lg n)$)
- Cuadrático ($O(n^2)$)
- Cúbico ($O(n^3)$)
- Exponencial ($O(a^n)$)

```python
import matplotlib.pyplot as plp
import numpy as np

def constante(x):
    return 2

def logaritmico(x):
    return np.log2(x)

def lineal(x):
    return x

def lineal_logaritmico(x):
    return x*np.log2(x)

def cuadratico(x):
    return x**2

def cubico(x):
    return x**3

def exponencial(x):
    return 2**x

x=np.arange(0.1,50,0.1)

plp.plot(x, [constante(i) for i in x], label="constante")
plp.plot(x, [logaritmico(i) for i in x], label="logaritmico")
plp.plot(x, [lineal(i) for i in x], label="lineal")
plp.plot(x, [lineal_logaritmico(i) for i in x], label="linearitmico")
plp.plot(x, [cuadratico(i) for i in x], label="cuadratico")
plp.plot(x, [cubico(i) for i in x], label="cubico")
plp.plot(x, [exponencial(i) for i in x], label="exponencial")

plp.xlim(0,50)
plp.ylim(0,50)

plp.legend(loc='best')

plp.show()
```

- **Progresión paso a paso de selección:**

```python
def swap(A,x,y):
    print("Posición "+str(x)+" por posición "+str(y))
    print(A)
    aux=A[x]
    A[x]=A[y]
    A[y]=aux
    
def selection(arr):
    n=len(arr)
    for i in range(0,n):
        minimum=i
        for j in range(i+1,n):
            if arr[j]<arr[minimum]:
                minimum=j
        swap(arr,i,minimum) 
        
arreglo=[32, 28, 63, 7, 73, 93, 98, 51, 26, 19, 2, 23, 46, 92, 3, 5, 88, 84, 81, 24]
selection(arreglo)
print ("Resultado final")
print (arreglo)
```

- **Progresión paso a paso para mezclas:**

```python
def mergesort(arr):
    print("Divide ",arr)
    if len(arr)>1:
        mid = len(arr)//2
        lefthalf = arr[:mid]
        righthalf = arr[mid:]

        mergesort(lefthalf)
        mergesort(righthalf)
        merge(arr,lefthalf,righthalf)
        print("Merge ", arr)
      
def merge(arr,lefthalf,righthalf):
    i=0
    j=0
    k=0
    while i < len(lefthalf) and j < len(righthalf):
        if lefthalf[i] < righthalf[j]:
            arr[k]=lefthalf[i]
            i=i+1
        else:
            arr[k]=righthalf[j]
            j=j+1
        k=k+1

    while i < len(lefthalf):
        arr[k]=lefthalf[i]
        i=i+1
        k=k+1

    while j < len(righthalf):
        arr[k]=righthalf[j]
        j=j+1
        k=k+1           
        
arreglo=[32, 28, 63, 7, 73, 93, 98, 51, 26, 19, 2, 23, 46, 92, 3, 5, 88, 84, 81, 24]
mergesort(arreglo)
print ("Resultado final")
print (arreglo)
```

- **Demostración a partir de la definición de cada notación asintótica usando límites.**

Sabemos que:

$$T(n)=\frac{n^2}{2}-\frac{n}{2}$$

En el caso de la notación **tilde**, por lo visto en clases sabemos que:

$$T(n)=\frac{n^2}{2}-\frac{n}{2}\sim\frac{n^2}{2}$$

A partir de su definición de límite, debemos demostrarlo:

$$\lim\limits_{n \to \infty}{\frac{f(n)}{g(n)}}=1$$
$$\lim\limits_{n \to \infty}{\frac{\frac{n^2}{2}-\frac{n}{2}}{\frac{n^2}{2}}}$$
$$2\lim\limits_{n \to \infty}{\frac{\frac{n^2}{2}-\frac{n}{2}}{n^2}}$$

Dividimos numerador y denominador por $n^2$:

$$2\lim\limits_{n \to \infty}{\frac{\frac{1}{2}-\frac{1}{2n}}{1}}$$
$$2\lim\limits_{n \to \infty}{\frac{1}{2}-\frac{1}{2n}}$$

Cuando $n$ tiende a infinito, $-\frac{1}{2n}$ tiende a $0$, por lo tanto:

$$2\lim\limits_{n \to \infty}{\frac{1}{2}}$$
$$2*\frac{1}{2}=1$$

***

Para el caso de **Big Oh**, por lo visto en case sabemos que:

$$T(n)=\frac{n^2}{2}-\frac{n}{2}=O(n^2)$$

A partir de su definición de límite, debemos demostrarlo:

$$\lim\limits_{n \to \infty}{\frac{f(n)}{g(n)}}< \infty$$
$$\lim\limits_{n \to \infty}{\frac{\frac{n^2}{2}-\frac{n}{2}}{n^2}}$$

Dividimos numerador y denominador por $n^2$:

$$\lim\limits_{n \to \infty}{\frac{\frac{1}{2}-\frac{1}{2n}}{1}}$$
$$\lim\limits_{n \to \infty}{\frac{1}{2}-\frac{1}{2n}}$$

Cuando $n$ tiende a infinito, $-\frac{1}{2n}$ tiende a $0$, por lo tanto:

$$\lim\limits_{n \to \infty}{\frac{1}{2}}$$
$$\frac{1}{2}<\infty$$

***

Para el caso de **little oh**:

$$T(n)=\frac{n^2}{2}-\frac{n}{2}=o(n^3)$$

A partir de su definición de límite, debemos demostrarlo:

$$\lim\limits_{n \to \infty}{\frac{f(n)}{g(n)}} = 0$$
$$\lim\limits_{n \to \infty}{\frac{\frac{n^2}{2}-\frac{n}{2}}{n^3}}$$

Dividimos numerador y denominador por $n^3$:

$$\lim\limits_{n \to \infty}{\frac{\frac{1}{2n}-\frac{1}{2n^2}}{1}}$$
$$\lim\limits_{n \to \infty}{\frac{1}{2n}-\frac{1}{2n^2}}$$

Cuando $n$ tiende a infinito, ambos argumentos tienden a $0$, por lo que se cumple.

- **Función que determina el algoritmo.**

Mi algoritmo descarta la idea inicial de encontrar un número, insertarlo en el arreglo y luego buscar el siguiente, verificar si esta en el arreglo, si no esta insertarlo, si esta buscar otro, y asi sucesivamente. Esto porque es tremendamente poco eficiente, siendo además muy dificil de predecir el final del algoritmo cuando $x-y$ es muy cercano o incluso $n$.

En cambio, se propone lo siguiente; crear un arreglo donde esten todos los numeros entre $[x...y]$, tomar una *posición* $i$ aleatoria del arreglo (con eso tenemos un numero entre $[x...y]$ aleatorio), eliminar la posición $i$ obtenida del arreglo, repetir hasta que tengamos los $n$ elementos. Con esto nos aseguramos que **NUNCA** hay colisiones, por ende no debemos comprobar en cada paso si el número existe. Como contrapartida, usamos mas espacio, pues debemos crear un arreglo extra de $y-x$ posiciones.

En el ejemplo, tenemos que el rango de números $[x...y]=[1...100]$, por lo que tenemos un universo de 100 posibles números que se guardan en *arr_x_y*. Además se nos piden de esos 100, $n=20$ números aleatorios, que se guardan en *arr_n*. Como se mencionó anteriormente, las colisiones son imposibles y el algoritmo funciona con la restricción que $y-x\le n$.

```python
import random as rnd

num_x = 1
num_y = 100
num_n = 20
arr_x_y = []
arr_x_y_size = 0
arr_n = []
index_i = 0

for i in range(num_x,num_y+num_x):
    arr_x_y.append(i)

print("UNIVERSO")
print(len(arr_x_y))
print(arr_x_y)

arr_x_y_size=len(arr_x_y)
for i in range(0,num_n):
    index_i=rnd.randint(0,arr_x_y_size)

    arr_n.append(arr_x_y[index_i])

    del arr_x_y[index_i]
    arr_x_y_size-=1

print("SAMPLE")
print(len(arr_n))
print(arr_n)
```

- **Análisis del algoritmo.**

El algoritmo esta determinado por la creación del arreglo *arr_x_y* y por la selección de los $n$ elementos aleatorios de *arr_n*, ambas son lineales a la cantidad de elementos. Si $m=y-x$, podemos deducir facilmente que nuestro algoritmo es $O(m+n)$, para todos los casos (Peor, Mejor y Promedio), siempre que *randint* se mantenga como $O(1)$, ya que siempre debe crear el arreglo y siempre debe encontrar $n$ elementos.

- **Que pasa si la función randint es $O(y-x)$**

Por la definción anterior $O(y-x)=O(m)$, si esto sucede, el segundo ciclo pasa a tener un costo de $O(n*m)$, es decir, el costo de cada elemento aleatorio, por el costo de conseguirlo. Esto nos daría un total general de $O(m+n*m)$ para el caso promedio. En este caso podemos definir un peor caso que es cuando $n \to m$, quedando el costo $O(m+m^2)=O(m^2)$, pasando de un costo lineal a uno cuadrático.
